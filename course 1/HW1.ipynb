```python
import numpy as np
matrix = np.zeros((n, d))
for i in range(n):
    for j in range(d):
        matrix[i][j] = data_list[i].count(unique_list[j])
```

```python
import math
def f(x):
    return math.sin(x / 5.) * math.exp(x / 10.) + 5. * math.exp(-x / 2.)
```

```python
f(1), f(15)
```

```python
x1 = [1, 15]
x2 = [1, 8, 15]
x3 = [1, 4, 10, 15]
```

```python
np.array(map(f, x3))
```

```python
def myA(dimA, x):
    return np.array([[x[i]**j for j in range(dimA)] for i in range(dimA)])
```

```python
myA(4, x3)
```

```python
import scipy
w1 = scipy.linalg.solve(myA(2,x1), np.array(map(f, x1)))
w2 = scipy.linalg.solve(myA(3,x2), np.array(map(f, x2)))
w3 = scipy.linalg.solve(myA(4,x3), np.array(map(f, x3)))
np.dot(myA(4,x3), w3)==np.array(map(f, x3))
```
